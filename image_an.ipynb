```python
def image_to_image(imgX,saved):
  img1 = Image.open(imgX).convert('L')
  img_ar = np.asarray(img1)
  img_df=pd.DataFrame(img_ar)

  # for i in img_df.columns:
  #   img_df[str(i)]=img_df[i]

  # new_df=pd.DataFrame()
  # '''edit image'''
  # for i in img.columns:


  
  # img_ar2=img_df.to_numpy()
  # new_img = Image.fromarray(img_ar2)
  # # print(new_img)
  # new_img.save(saved)

  for i in img_df.columns:
    img_df[str(i)]=img_df[i]
    img_df.drop(i,inplace=True)

  new_df=pd.DataFrame()

  for i in img_df.columns:

    ARIMA_MODEL = sm.tsa.statespace.SARIMAX(img_df[i][:800])
                                              
                                              
      

      # Fit the model and print results
      
    results = ARIMA_MODEL.fit()
      
      #predict values 18 months out
    pred_value=results.predict(start=0,end=len(img_df[0])-1)
    new_df[i]=pred_value
  # return new_df
  img_ar2=new_df.to_numpy()
  new_img = Image.fromarray(img_ar2)
  # print(new_img)
  new_img.save(saved)  
```
